```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Question 1(A) part 1

```python
c0 = 6
k = [0.1386,0.2,0.3]
t = np.arange(0, 24, 1)


plt.figure(1)
for k_curr in k:
    plt.plot(t, c0*np.exp(-k_curr*t), label = 'k ='+ str(k_curr))
    t_half = 0.693/k_curr
    thalf = int(t_half)
    plt.scatter(thalf, c0*np.exp(-k_curr*thalf))
    plt.axvline(t_half, linestyle='dashed',color= 'black')
plt.title('Conc. decay')
plt.xlabel('Time(in hours)')
plt.ylabel('Concentration')
plt.legend()
plt.grid()
plt.savefig('q1a_part1.png')
```

# Question 1(A) part 2

```python
%matplotlib inline
k = [0.1386,0.2,0.3]
t = np.linspace(0, 24, 24*60)
time_until = 5 #in hours
plt.figure(1)

for k_curr in k:
    y = []
    count = 0
    c0 = 6/(time_until*60)
    for t_curr in range(t.shape[0]):
        if(t[t_curr]<(time_until)): 
            c0 = c0+6/(time_until*60);
            count = count+1
        y.append(c0*np.exp(-k_curr*t[t_curr]))
    plt.plot(t, y, label = 'k ='+ str(k_curr))
    
    t_half = 0.693/k_curr
    thalf = int(t_half)
    plt.scatter(thalf, y[thalf*60])
    plt.axvline(thalf, linestyle='dashed',color= 'black')

plt.axvline(time_until,color= 'red')
plt.title('Conc. decay(6 units upto 5 hours)')
plt.xlabel('Time(in hours)')
plt.ylabel('Concentration')
plt.legend()
plt.grid()
plt.savefig('q1a_part2.png')

#########################################
#What if it was more or less than 5 hours
k = [0.1386]
t = np.linspace(0, 24, 24*60)
plt.figure(2)

time_until = 5 #in hours

for k_curr in k:
    y = []
    count = 0
    c0 = 6/(time_until*60)
    for t_curr in range(t.shape[0]):
        if(t[t_curr]<(time_until)): 
            c0 = c0+6/(time_until*60);
            count = count+1
        y.append(c0*np.exp(-k_curr*t[t_curr]))
    plt.plot(t, y, label = '5 hour',color = 'blue')
    
    t_half = 0.693/k_curr
    thalf = int(t_half)
    plt.scatter(thalf, y[thalf*60])
plt.axvline(time_until,color= 'red')

    

time_until = 1

for k_curr in k:
    y = []
    count = 0
    c0 = 6/(time_until*60)
    for t_curr in range(t.shape[0]):
        if(t[t_curr]<(time_until)): 
            c0 = c0+6/(time_until*60);
            count = count+1
        y.append(c0*np.exp(-k_curr*t[t_curr]))
    plt.plot(t, y, label = '1 hour',color = 'green')
    t_half = 0.693/k_curr
    thalf = int(t_half)
    plt.scatter(thalf, y[thalf*60])
    

plt.axvline(time_until,color= 'red')
plt.title('Conc. decay(6 units upto 1 and 5 hours )')
plt.xlabel('Time(in hours)')
plt.ylabel('Concentration')
plt.legend()
plt.grid()
plt.savefig('q1a_part3.png')
```
